###Data Preprocessing
   Data pre-processing is an essential step in any data
   analysis work flow.it ensures data is clean,consistent,
   and ready for deeper analysis.

```python
#remove duplicates
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
#detecting null values
df.isnull().sum()
```

###Correcting Data Types
converted columns to appropriate types for better analysis.

```python
df["Month"] = df["Month"].astype("category")
df["Weekend"] = df["Weekend"].astype(bool)
df["Revenue"] = df["Revenue"].astype(bool)
df.info()
```

###Creating Derived columns


```python
#Total pages visited
df["TotalPages"] = (
    df["Administrative"] +
    df["Informational"] +
    df["ProductRelated"]
)
df
```

```python
#Total duration spent
df["TotalDuration"] = (
    df["Administrative_Duration"] +
    df["Informational_Duration"] +
    df["ProductRelated_Duration"]
)
df
```

###Filtering

```python
#filter sessions that generated revnue
revenue_sessions = df[df["Revenue"] == True]
revenue_sessions
```

```python
#filter returning visitors
returning_visitors = df[df["VisitorType"] == "Returning_Visitor"]
returning_visitors
```

```python
#filter weekend sessions
weekend_sessions = df[df["Weekend"] == True]
weekend_sessions
```

```python
#filter low bounce rate
low_bounce = df[df["BounceRates"] < 0.05]
df
```

```python
#sessions that generated revenue
buyers = df[df["Revenue"] == True]
buyers
```

```python
#sessions without revenue
non_buyers = df[df["Revenue"] == False]
non_buyers
```

###Encoding Categorical Variables


*   converts categorical data into numerical format  



```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df["Month"] = le.fit_transform(df["Month"])
df["VisitorType"] = le.fit_transform(df["VisitorType"])

df["VisitorType"]
```

```python
df["Month"]
```

###Correlation Analysis

*   identifies redundant features



```python
corr_matrix = df.corr(numeric_only=True)

high_corr = corr_matrix.abs() > 0.9
high_corr
```
